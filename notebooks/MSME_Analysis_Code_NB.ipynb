```python
mean_val = df33['Credit_Lakh_Msme'].mean()
median_val = df33['Credit_Lakh_Msme'].median()

print(f"National Mean: {mean_val:.2f}")
print(f"National Median: {median_val:.2f}")
```

```python
print(df33['Total MSME'].median())
```

```python

benchmark = 3.84

df33['Gap_Percentage'] = ((df33['Credit_Lakh_Msme'] - benchmark) / benchmark) * 100
df33['Gap_Percentage'] = df33['Gap_Percentage'].round(2)
```

```python
df33
```

```python
df33
```

```python
import numpy as np
df33['Credit_Lakh_Msme'] = df33['Credit_Lakh_Msme'].replace([np.inf, -np.inf], 0)
df33
```

```python
#to assign priority that where should a bank look up to
def assign_priority(row):
    if row['Gap_Percentage'] < -50 and row['Is_Aspirational'] == True:
        return "1. Critical (Aspirational Desert)"
    elif row['Gap_Percentage'] < -50:
        return "2. High Gap"
    elif row['Gap_Percentage'] < 0:
        return "3. Underfunded"
    else:
        return "4. Stable"

df33['Priority_Level'] = df33.apply(assign_priority, axis=1)
```

```python
print(df33['Priority_Level'].value_counts())
```
